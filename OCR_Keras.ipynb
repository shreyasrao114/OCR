```
import imutils
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pafy
from IPython.display import YouTubeVideo
url = 'https://www.youtube.com/watch?v=TB_58npl9_o'
video = pafy.new(url)
best = video.getbest(preftype="mp4")
vidcap = cv2.VideoCapture(best.url)
```

```
from google.colab.patches import cv2_imshow
!pip install keras-ocr
```

```
import keras_ocr
pipeline = keras_ocr.pipeline.Pipeline()
```

```
size = (int(vidcap.get(3)), int(vidcap.get(4)))
result = cv2.VideoWriter('result.mp4', cv2.VideoWriter.fourcc(*'MPEG'), 60, size)
count=0
while(count<2000): #taking only 2000 frames to prevent disconnect
    ret, frame = vidcap.read()
    if ret == True:
      print('frame ' + str(count+1))
      count+=1
      words = []  #list of detected words
      gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
      predictions = pipeline.recognize([frame])[0]
      frame = keras_ocr.tools.drawBoxes(image=frame, boxes=predictions,thickness=1, boxes_format='predictions')
      for pred in predictions:
        words.append(pred[0])
      print(words)
      result.write(frame)
result.release()
vidcap.release()
```
